```python
import pandas as pd
df_farfetch = pd.read_csv('C:\\Users\\BGCNHK\\Desktop\\farfetch\\farfetch_selenium_1.csv', header=None)
df_farfetch.columns = ['price','brand','subtitle','description', 'made-in','material', 'style-id']
df_farfetch['brand'] = df_farfetch['brand'].str.lower()
df_farfetch['price'] = df_farfetch['price'].str.replace('^[^\d]', '').str.replace(',','').astype(float)
df_farfetch
```

```python
df_saks = pd.read_csv('C:\\Users\\BGCNHK\\Desktop\\saks\\saks.csv')
df_saks['product_titles'] = df_saks['product_titles'].str.lower()
df_saks
```

```python
df_combine = df_saks.groupby('product_titles',as_index=False)[['price','color','sub_title']].agg(['mean','nunique','count'])
# df_color = df_saks.groupby('product_titles',as_index=False)[['color']].count()
df_combine.columns = ['Average Price', 'Unqiue Color','Total Products']
df_combine.reset_index(level=0, inplace=True)
df_combine.columns.values[0] = 'brand'
df_combine
```

```python
df_combine_ff = df_farfetch.groupby('brand',as_index=False)[['price','subtitle']].agg(['mean','count'])
# df_combine_ff[['brand']].column = ['product_titles']
df_combine_ff.columns = ['Average Price','Total Products']
df_combine_ff.reset_index(level=0, inplace=True)
df_combine_ff
```

```python
df_merge = df_combine.merge(df_combine_ff,on='brand', how='left')
df_merge
```

```python
df_merge = df_merge.dropna(axis=0, how='any')
df_merge.columns = ['brand', 'Saks Avg Price', 'Saks Unique Color', 'Saks Total Products','Farfetch Avg Price','Farfetch Total Products']
df_merge['Farfetch Total Products'] = df_merge['Farfetch Total Products'].astype(int)
df_merge['Farfetch Avg Price'] = df_merge['Farfetch Avg Price'].astype(int)
df_merge['Saks Avg Price'] = df_merge['Saks Avg Price'].astype(int)
df_merge.copy = df_merge.sort_values(by='Saks Avg Price')
df_merge.copy
```
